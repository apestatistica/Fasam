```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
# data = {
#     'x':[1,2,1,2,8,9,9,8,1,2,1,2],
#     'y':[2,1,1,2,9,8,9,8,15,15,14,14]
# }
# 
# df = pd.DataFrame.from_dict(data)

df = pd.DataFrame.from_dict(np.random.randint(100, size=(50, 2)))
df.columns = ['x','y']

df
```

```python
# cls = KMeans(n_clusters=3, random_state=0).fit(df)
cls = KMeans(n_clusters=3).fit(df)
predictions = cls.predict(df)
print('Clusters:')
print(pd.Series(cls.labels_).value_counts())
```

```python
x, y = [], []
for i in range(df.shape[0]):
    x.append(df['x'][i])
    y.append(df['y'][i])

plt.scatter(x, y, c=cls.labels_, cmap='viridis')
for i in cls.cluster_centers_:
    plt.plot(i[0], i[1], 'x', color='red')
```
